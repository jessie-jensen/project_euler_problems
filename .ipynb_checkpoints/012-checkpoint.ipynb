#Problem 12: Highly divisible triangular number

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be $ 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28 $. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:
+ 1: 1
+ 3: 1,3
+ 6: 1,2,3,6
+ 10: 1,2,5,10
+ 15: 1,3,5,15
+ 21: 1,3,7,21
+ 28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
def triangle_nums (list_length):
    '''function that returns a list of triangle numbers of length list_length'''
    triangles = []
    num = 0

    for i in xrange(1, list_length + 1):
        num += i
        triangles.append(num)

    return triangles

# test function
triangle_nums(7)
```

```python
def factorization (x):
    '''function that returns a list of all factors of x'''
    factors = []
    i = 1
    check = x
    
    while (i <= x/i):
        if x % i == 0:
            factors.append(i)
            factors.append(x/i)
            check = x/i
        i += 1
    
    factors.sort()
    
    return factors

# test function
factorization(28)
```

```python
def triangle_num_factors_threshold (threshold, list_max = 1000*1000):
    '''blah'''
    triangles = triangle_nums(list_max)
    
    i = 0
    factors = 0
    
    while factors < threshold:
        factors = len(factorization(triangles[i]))
        i += 1
        
    return triangles[i-1]
        
# test function
triangle_num_factors_threshold(5)
```
